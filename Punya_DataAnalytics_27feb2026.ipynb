```python
data = pd.DataFrame({
    'name': ['John', 'Jane', 'Jack', 'John', None],
    'age': [28, 34, None, 28, 22],
    'purchase_amount': [100.5, None, 85.3, 100.5, 50.0],
    'date_of_purchase': ['2023/12/01', '2023/12/02', '2023/12/01', '2023/12/01', '2023/12/03']
})

data
```

Data Transformation :- Creating Dataset

```python
data = pd.DataFrame({
    'category': ['A', 'B', 'A', 'C', 'B'],
    'numeric_column': [10, 15, 10, 20, 15]
})

data
```

Scale Numerical Data

```python
scaler = StandardScaler()
data['scaled_numeric_column'] = scaler.fit_transform(
    data[['numeric_column']]
)

data
```

One-Hot Encode Categorical Data

```python
encoder = OneHotEncoder(sparse_output=False)

encoded_data = pd.DataFrame(
    encoder.fit_transform(data[['category']]),
    columns=encoder.get_feature_names_out(['category'])
)

data = pd.concat([data, encoded_data], axis=1)
data
```

Data Reduction :- Create Dataset

```python
data = pd.DataFrame({
    'feature1': [10, 20, 30, 40, 50],
    'feature2': [1, 2, 3, 4, 5],
    'feature3': [100, 200, 300, 400, 500],
    'target': [0, 1, 0, 1, 0]
})
data
```

Feature Selection (SelectKBest)

```python
selector = SelectKBest(score_func=chi2, k=2)
selected_features = selector.fit_transform(
    data[['feature1', 'feature2', 'feature3']],
    data['target']
)

print("Selected Features:")
print(selected_features)
```

Dimensionality Reduction (PCA)

```python
pca = PCA(n_components=2)
pca_data = pca.fit_transform(
    data[['feature1', 'feature2', 'feature3']]
)

print("PCA Output:")
print(pca_data)
```

Handling Missing Data (Techniques)

```python
data = pd.DataFrame({
    'column_with_missing': [10, None, 30, None, 50]
})

imputer = SimpleImputer(strategy='mean')
data['column_with_missing'] = imputer.fit_transform(
    data[['column_with_missing']]
)

print(data)
```

Deletion

```python
data = pd.DataFrame({
    'A': [1, None, 3],
    'B': [4, 5, None]
})

data.dropna(inplace=True)
print(data)
```

Outlier Detection :- Z-Score Method

```python
data = pd.DataFrame({'column': [10, 12, 14, 100, 13]})

z_scores = np.abs(stats.zscore(data['column']))
outliers = z_scores > 3

print("Outliers detected:")
print(data[outliers])
```

IQR Method

```python
Q1 = data['column'].quantile(0.25)
Q3 = data['column'].quantile(0.75)
IQR = Q3 - Q1

outliers = data[
    (data['column'] < Q1 - 1.5 * IQR) |
    (data['column'] > Q3 + 1.5 * IQR)
]

print(outliers)
```

Data Encoding :- One-Hot Encoding

```python
data = pd.DataFrame({'color': ['red', 'blue', 'green']})

encoder = OneHotEncoder(sparse_output=False)
encoded = encoder.fit_transform(data[['color']])

print(encoded)
```

```python
le = LabelEncoder()
data['color_encoded'] = le.fit_transform(data['color'])

print(data)
```

Ordinal Encoding

```python
data = pd.DataFrame({'size': ['low', 'medium', 'high']})

oe = OrdinalEncoder(categories=[['low', 'medium', 'high']])
data['size_encoded'] = oe.fit_transform(data[['size']])

print(data)
```

Scaling & Normalization :- Min-Max Scaling

```python
data = pd.DataFrame({'value': [10, 20, 30]})

scaler = MinMaxScaler()
data['scaled'] = scaler.fit_transform(data[['value']])

print(data)
```

Standardization

```python
scaler = StandardScaler()
data['standardized'] = scaler.fit_transform(data[['value']])

print(data)
```

Preprocessing Pipeline (FINAL)

```python
import numpy as np
data = pd.DataFrame({
    'age': [25, np.nan, 30],
    'category': ['A', 'B', np.nan]
})

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer

numeric_pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())
])

categorical_pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('encoder', OneHotEncoder(handle_unknown='ignore'))
])

preprocessor = ColumnTransformer(transformers=[
    ('num', numeric_pipeline, ['age']),
    ('cat', categorical_pipeline, ['category'])
])

processed_data = preprocessor.fit_transform(data)
print(processed_data)
```
